```python
def preprocess(df):
    # dtype transformation:
    df = df.apply(lambda col: col.astype(float) if col.dropna().nunique() > 10 \
                 else col.astype(int))
    df['APOE4'] = df['APOE4'].astype(float)
    df, mean_sd = dc.standardization(df)
    return df
```

```python
train_dct = {}
test_dct = {}
for i in range(1,6):
    train = pd.read_csv(f"../../clean_data/impute/train{i}.csv")
    test = pd.read_csv(f"../../clean_data/impute/test{i}.csv")
    train_dct[i] = preprocess(df=train)
    test_dct[i] = preprocess(df=test)
dc.save_py(train_dct, "../../clean_data/train_set")
dc.save_py(test_dct, "../../clean_data/test_set")
```
